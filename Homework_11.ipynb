```python
import matplotlib.pyplot as plt
import sympy
from sympy import *
from random import randint
import math
```

```python
RANGEE = 100
acc = 2
ACCURACY = 10**acc
func = lambda x : -12*x**4*math.sin(math.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
points = []
for i in range(RANGEE * ACCURACY + 1):
    points.append(((-RANGEE * ACCURACY / 2) + i) / ACCURACY)
print('Start = ', points[0], '\nEnd = ', points[len(points) - 1])
points = list(map(func, points))
plt.plot(points)
```

```python
# 1, 6, 7. Корни уравненияц, f > 0, f < 0
root = []
ower = []
sub = []
start_f = -RANGEE / 2
end_f = 0
for i in range(1, len(points)):
    if (points[i - 1] >= 0 and points[i] <= 0) or (points[i - 1] <= 0 and points[i] >= 0):
        root.append(round((i - 1) / ACCURACY -RANGEE / 2, acc))
        if points[i - 1] > 0:
            end_f = (round((i - 1) / ACCURACY -RANGEE / 2, acc))
            a = (start_f, end_f)
            ower.append(a)
            start_f = (round(i / ACCURACY -RANGEE / 2, acc))
        else:
            end_f = (round((i - 1) / ACCURACY -RANGEE / 2, acc))
            a = (start_f, end_f)
            sub.append(a)
            start_f = (round(i / ACCURACY -RANGEE / 2, acc))

print('f > 0\n', ower, '\n\nf < 0:\n', sub)            
print('\n\nФункция на промежутке [', RANGEE / -2, ':', RANGEE / 2, '] имеет значения равные 0 в точках:\n',\
      root, '\nС точностью до', ACCURACY, ' символов')
```

```python
# Жаль что местная библиотека не справляется с дифиринциированием такой функции. Пришлось другим путем ее дифиринциировать 
dif_func = lambda x: 12*x**4*sin(x)*cos(cos(x))-48*x**3*sin(cos(x))-54*x**2+10*x+10
dif_points = []
for i in range(RANGEE * ACCURACY + 1):
    dif_points.append(((-RANGEE * ACCURACY / 2) + i) / ACCURACY)
dif_points = list(map(dif_func, dif_points))
plt.plot(dif_points)
```

```python
# 5, 2, 3. Вершины Убывание, Возрастание 

apex = []
increes = []
reduse = []
start_f = -RANGEE / 2
end_f = 0
for i in range(1, len(points)):
    if (dif_points[i - 1] >= 0 and dif_points[i] <= 0) or (dif_points[i - 1] <= 0 and dif_points[i] >= 0):
        apex.append(round((i - 1) / ACCURACY -RANGEE / 2, acc))
        if points[i - 1] > 0:
            end_f = (round((i - 1) / ACCURACY -RANGEE / 2, acc))
            a = (start_f, end_f)
            increes.append(a)
            start_f = (round(i / ACCURACY -RANGEE / 2, acc))
        else:
            end_f = (round((i - 1) / ACCURACY -RANGEE / 2, acc))
            a = (start_f, end_f)
            reduse.append(a)
            start_f = (round(i / ACCURACY -RANGEE / 2, acc))

print('Функция возрастает на промежутках\n', increes, '\n\nФункция убывает на промежутках:\n', reduse)            
print('\n\nФункция на промежутке [', RANGEE / -2, ':', RANGEE / 2, '] имеет вершиы в точках:\n',\
      apex, '\nС точностью до', ACCURACY, ' символов')
```
